# A. Generate a synthetic dataset where y=4x+5+ε, with ϵ∽N(0,1). Use 100 samples.

```python
# Nguyen Andy
# SWE3050-41
# 성균관대학교 SKKU

import numpy as np
import pandas as pd

# set seed for reproducibility
np.random.seed(1234)

# generate 100 samples for x
x = np.random.rand(100, 1) * 10 # generate random x values between 0 and 10

# generate noise from a normal distribution with mean 0 and standard deviation 1
epsilon = np.random.normal(0, 1, size=(100, 1))

# generate y based on the given formula
y = 4 * x + 5 + epsilon

# create a pandas DataFrame to store the dataset
data = pd.DataFrame({'x': x.flatten(), 'y': y.flatten()})

# display the first few rows of the dataset
display(data.head())
```

# B. Fit a linear regression model to this dataset.
# C. Show the learned coefficients (slope and intercept)

```python
# Nguyen Andy
# SWE3050-41
# 성균관대학교 SKKU
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# generate data
np.random.seed(1234)
samples = 100
x = np.random.rand(samples, 1) * 10  # predictor
epsilon = np.random.normal(0, 1, (samples, 1))
y = 4 * x + 5 + epsilon

# fit linear regression
model = LinearRegression()
model.fit(x, y)  # x must be 2D, y can be 1D or 2D

print("Estimated slope (coefficient):", model.coef_[0][0])
print("Estimated intercept:", model.intercept_[0])

# predict on x
y_pred = model.predict(x)

# show first few predicted values
print(pd.DataFrame({'x': x.flatten(), 'y_true': y.flatten(), 'y_pred': y_pred.flatten()}).head())
```

# E. Plot the data points and the fitted regression line in a drawing tool (Matplotlib)

```python
# Nguyen Andy
# SWE3050-41
# 성균관대학교 SKKU
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# generate data
np.random.seed(1234)
n_samples = 100
x = np.random.rand(n_samples, 1) * 10  # predictor
epsilon = np.random.normal(0, 1, (n_samples, 1))
y = 4 * x + 5 + epsilon

# fit linear regression=
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)

# plot
plt.figure(figsize=(10, 8))

# actual data scatter points
plt.scatter(x, y, color='blue', alpha=0.6, label='data points')

# regression line — sort x to get a smooth line
sorted_idx = np.argsort(x.flatten())
plt.plot(x.flatten()[sorted_idx],
         y_pred.flatten()[sorted_idx],
         color='red', linewidth=2, label='fitted regression line')

plt.title('Linear Regression Fit: y = 4x + 5 + ε')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
